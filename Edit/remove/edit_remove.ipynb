```python
import pandas as pd
import numpy as np
import os, re

df = pd.read_csv(r'C:\GOD\programming\NLP\DecoLexO\Edit\example\DecoLex_Test.csv')


####행 이름 decolex처럼 변경하기####

#첫 행 살리기
col_rgx = re.compile(r'(.*)')
first = list(df.columns)
df.loc[0] = first
for val in first:
    m = col_rgx.match(val)
    if m:
        x = first.index(val)
        first[x] = np.nan
    else:
        continue
df.loc[0] = first

#행 이름 설정해주기
col_nme = ['Lemma','Category','Morph1']

l = len(df.columns)
for i in range(1, l - 2):
    info = 'info' + str(i)
    col_nme.append(info)

df.columns = col_nme
print(df)
```
