```python
##begin을 replace

res = []
ori = input('original')
new = input('new')
for x in lemma:
    y = re.sub('^'+ori,new,x)
    res.append(y)
    
print(res)
    #열 추가
# del df['Lemma']
df['Lemma'] = res

#열 위치 재정렬
df = df[col_nme]
print(df)

#함수화

def begin_rpl(df, col, original_text, new_text):
    res = []
    first = df2first(df,col)
    #begin에서 replace
    for x in first:
        y = re.sub('^' + original_text, new_text, x)
        res.append(y)
        
    #열 추가
    df[col] = res
    
    #열 위치 재정렬
    df = df[col_nme]
    
    print(df)
```

```python
##end를 replace

res = []
ori = input('original')
new = input('new')
for x in lemma:
    y = re.sub(ori + '$',new,x)
    res.append(y)
    
print(res)
    #열 추가
# del df['Lemma']
df['Lemma'] = res

#열 위치 재정렬
df = df[col_nme]
print(df)

#함수화

def end_rpl(df, col, original_text, new_text):
    res = []
    first = df2first(df,col)
    #end에서 replace
    for x in first:
        y = re.sub(original_text + '$', new_text, x)
        res.append(y)
        
    #열 추가
    df[col] = res
    
    #열 위치 재정렬
    df = df[col_nme]
    
    print(df)
```

```python
def rpl(df, col, original_text, new_text, where):
    if where == 'anywhere':
        anywhere_rpl(df, col, original_text, new_text)
    if where == 'whole string':
        whole_rpl(df, col, original_text, new_text)
    if where == 'begin':
        begin_rpl(df, col, original_text, new_text)
    if where == 'end':
        end_rpl(df, col, original_text, new_text)
    
rpl(df,col, ori, new, input('where? '))
```
